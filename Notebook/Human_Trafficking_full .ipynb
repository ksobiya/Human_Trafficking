```python
#creating a forloop and adding a column to categorize each exploitation type and combining nan and other together

for index, row in h_t.iterrows():
    if row['Exploitation Type']=='Sexual exploitation':
        h_t.loc[index,'exp_type']='Sexual exploitation'
    elif row['Exploitation Type']=='Forced labour':
        h_t.loc[index,'exp_type']='Forced labour'
    elif row['Exploitation Type']=='Forced labour;Other':
        h_t.loc[index,'exp_type']='Forced labour'
    elif row['Exploitation Type']=='Forced labour;Sexual exploitation;Combined sexual and labour exploitation':
        h_t.loc[index,'exp_type']='Combined sexual and labour exploitation'
    elif row['Exploitation Type']=='Forced marriage':
        h_t.loc[index,'exp_type']='Forced marriage'
    else:
        h_t.loc[index,'Exploitation Type'] = 'Other'
```

```python
# dropna will give all the NAN values too.... #The size() method is applied to the grouped data. It returns a Series with the number of occurrences of each unique combination of values in the specified grouping columns. 
#In other words, it counts the number of rows in each group. or adds the rows in each group.
exp_df.groupby(['Country_of_Exploitation','exp_type'], dropna=False).size().reset_index(name='count')
```

```python
# this will not include nan values
# same way of doing the above instead in this line of code we are using .size() that is counding the number of rows in each grp

countries_of_exp= exp_df.groupby(['Country_of_Exploitation','exp_type']).size().reset_index(name='count')
countries_of_exp
```

```python
# Pivot the DataFrame to have 'Country_of_Exploitation' as index and 'exp_type' as columns

pivot_df=countries_of_exp.pivot(index='Country_of_Exploitation', columns='exp_type', values='count').fillna(0)
pivot_df
```
